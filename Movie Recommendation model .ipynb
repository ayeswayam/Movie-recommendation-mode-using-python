```python
# Importing the required library
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Data Collection and Pre-Processing
movies_data = pd.read_csv('C:\\Users\\ayesw\\Desktop\\movie prediction project\\movies.csv')
# printing the data to check the loading process.
movies_data.head()
```

```python
# to find out the number of rows and columns in the data frame
movies_data.shape
```

```python
# to select the relevant features for recommendation
relevant_features = ['genres','keywords','tagline','cast','director']
print(relevant_features)
```

```python
# replacing the null valuess with null string
for feature in relevant_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
# combining all the 5 relevant features
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
print(combined_features)
```

```python
# converting the text data to feature vectors
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

```python
## Cosine Similarity
# getting the similarity scores using cosine similarity
similarities = cosine_similarity(feature_vectors)
print(similarities)
#getting numbers of rows and columns of similarity shapes
print("similarities shapes")
print(similarities.shape)
```

```python
# getting the movie name from the user
movie_name_from_user = input(' Enter your favourite movie name : ')
```

```python
# creating a list with all the movie names given in the dataset
list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
# finding the close match for the movie name given by the user
finding_close_match = difflib.get_close_matches(movie_name_from_user, list_of_all_titles)
print(finding_close_match)
close_match = finding_close_match[0]
print(close_match)
```

```python
# finding the index of the movie with title
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# getting a list of similar movies
similarity_score = list(enumerate(similarities[index_of_the_movie]))
print(similarity_score)
len(similarity_score)
```

```python
# sorting the movies based on their similarity score
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index
print('Movies suggested for you : \n')
i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```

```python
#movie recommendation system
movie_name = input(' Enter your favourite movie name : ')
list_of_all_titles = movies_data['title'].tolist()
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
close_match = find_close_match[0]
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
similarity_score = list(enumerate(similarities[index_of_the_movie]))
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print('Movies suggested for you : \n')
i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
